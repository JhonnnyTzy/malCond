```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos las ecuaciones del sistema
# 1.0001x + 1.0000y = 2
# 1.0000x + 1.0000y = 2

# Creamos los puntos de x
x = np.linspace(0, 2, 100)

# Despejamos y para cada ecuacin
y1 = (2 - 1.0001 * x) / 1.0000
y2 = (2 - 1.0000 * x) / 1.0000

# Creamos la grfica
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label="1.0001x + 1.0000y = 2", color='blue')
plt.plot(x, y2, label="1.0000x + 1.0000y = 2", color='red', linestyle='--')

# Aadimos etiquetas y ttulo
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grfico de las ecuaciones lineales')

# Aadimos una leyenda
plt.legend()

# Mostramos la grfica
plt.grid(True)
plt.show()
```
